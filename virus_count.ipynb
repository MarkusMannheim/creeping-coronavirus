```python
import pandas as pd
data = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vQguJUu1eZrH057CfqZpQnti5ESp6-qU3pDcYXak_Y-ZT9AJFN0j8KXJsoU_JapILVykJcLL0719dFr/pub?gid=0&single=true&output=csv")
data.fillna(0, inplace=True)

def cumulate(list):
    i = 0
    while i < len(list):
        if i != 0:
            list[i] = list[i] + list[i - 1]
        i = i + 1
    return list

for column in data.columns[1:]:
    data[column] = cumulate(list(data[column]))
data.to_csv("virusData.csv", index=False)
data
```
